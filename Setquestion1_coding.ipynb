```python
#5.Write a Python program to check if a number is prime. Expected Output (for input 7):
#7 is a prime number
import math
def prime_num(num1):
    if (num1 <= 1):
        return False
    if (num1 == 2):
        return True
    if ((num1 % 2) == 0):
        return False
    for i in range(3, int(math.sqrt(num1)) + 1, 2):
        if ((num1 % i) == 0):
            return False
    return True
```

```python
number = 7
if prime_num(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```
